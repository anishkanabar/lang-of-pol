Looks like a lot of the misses are thanks to typos, adding/removing "the", etc. There are only 39 of them so it'll be faster for me to correct by hand rather than using a fuzzy matching package.

```python
location_errors = {
    "Arie Crown Theater*": "Arie Crown Theater",
    "Cinespace Chicago Film Studio":"Cinespace Chicago Film Studios",
    "Civic Opera House (Lyric Opera of Chicago)":"Civic Opera House",
    "Credit Union 1 Arena (previously UIC Pavilion)":"Credit Union 1 Arena",
    "McCormick Place (Skyline Ballroom)*":"McCormick Place",
    "Millenium Park / Jay Pritzker Pavilion":"Millennium Park",
    "Petrillo Music Shell (inside Grant Park)":"Petrillo Music Shell",
    "Wintrust Arena*":"Wintrust Arena"
}

for error in location_errors:
    locations.loc[locations["Venue Name"] == error, "Venue Name"] = location_errors[error]
```

```python
event_errors = {
    "Guarenteed Rate Field": "Guaranteed Rate Field",
    "House of Blues - Chicago": "House of Blues",
    "Jay Pritzker Pavilion - Millennium Park": "Millennium Park",
    "Petrillo Music Shell (Grant Park)": "Petrillo Music Shell",
    "THalia Hall": "Thalia Hall",
    "The Field Museum":"Field Museum",
    "The Metro":"Metro",
    "The Riviera":"Riviera Theatre",
    "The Vic Theatre":"Vic Theatre",
    "the Chicago Theatre":"The Chicago Theatre"
}

for error in event_errors:
    events.loc[events["Location"] == error, "Location"] = event_errors[error]
```

```python
# now repeat!
geocoded_events = pd.merge(events,locations,how='left',left_on='Location',right_on='Venue Name')
geocoded_events.head()
```
